Escolha um ponto como origem para o sistema de coordenadas da câmera.

b. Transforme os objetos do sistema de coordenadas do mundo para o sistema de coordenadas da câmera.

d. Coloque um ponto informando a origem do sistema de coordenadas do mundo.

# Item A

```python
import numpy as np

# a) Defina origem da câmera, ponto de interesse e vetor up

# Calcula o centróide (ponto médio) dos objetos na cena
# Ignora objetos sem faces (como linhas) para o cálculo do centróide
medias_objetos = [
    np.mean(item[1], axis=0) for item in objetos if item[2] is not None
]

# Calcula o centróide geral a partir das médias dos objetos
centroide = np.mean(medias_objetos, axis=0)

# Define a origem da câmera (C) e o ponto de interesse (P) com base no centróide
# A câmera é posicionada acima do centróide (adicionando [0, 0, 10] ao centróide)
C = centroide + np.array([0, 0.0, 10])
# O ponto de interesse é o próprio centróide
P = centroide
# Define o vetor 'up' do mundo (eixo Z positivo)
up_world = np.array([0.0, 0.0, 1.0])

# Imprime as informações da câmera e do vetor up do mundo
print(f"Origem da câmera (C): {C}")
print(f"Ponto de interesse (P): {P}")
print(f"Vetor up do mundo: {up_world}")

# b) Base vetorial ortonormal da câmera
# Calcula o vetor de visão (w), apontando da câmera para o ponto de interesse
w = (C - P)
# Normaliza o vetor w
w = w / np.linalg.norm(w)

# Verifica se o vetor w é quase paralelo ao vetor up do mundo
# Se for, usa um vetor up alternativo para evitar problemas com o cross product
if np.abs(np.dot(w, up_world)) > 0.98:
    up_world = np.array([0.0, 1.0, 0.0]) # Usa o eixo Y como up alternativo

# Calcula o vetor 'right' (u) usando o cross product do vetor up do mundo e w
u = np.cross(up_world, w)
# Normaliza o vetor u
u = u / np.linalg.norm(u)

# Calcula o vetor 'up' da câmera (v) usando o cross product de w e u
v = np.cross(w, u)

# Imprime a base vetorial do sistema da câmera
print("\nBase vetorial do sistema da câmera:")
print("u (right):", u)
print("v (up):   ", v)
print("w (view): ", w)

# Matriz de rotação (para o item B)
# Cria a matriz de rotação cujas linhas são os vetores u, v, w
R = np.stack([u, v, w], axis=0)
```

# Item B, C, D

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Define a origem da câmera (C), o ponto de interesse (P) e o vetor 'up' do mundo
# C, P e up_world são calculados em uma célula anterior, usando o centróide dos objetos.
C = np.array(C)              # Origem da câmera (posição)
P = np.array(P)              # Ponto para onde a câmera está olhando
up_world = np.array(up_world) # Vetor 'para cima' no sistema do mundo

# Base vetorial ortonormal da câmera (u, v, w)
u = np.array(u)              # Vetor 'para a direita' da câmera
v = np.array(v)              # Vetor 'para cima' da câmera
w = np.array(w)              # Vetor de visão (apontando da câmera para o ponto de interesse)

# Matriz de rotação do sistema do mundo para o sistema da câmera
# As linhas da matriz são os vetores u, v, w.
R = np.stack([u, v, w], axis=0)

# Função para transformar vértices do sistema do mundo para o sistema da câmera
def world_to_camera(V):
    """
    Transforma um conjunto de vértices do sistema de coordenadas do mundo
    para o sistema de coordenadas da câmera.

    Args:
        V (np.ndarray): Array (N, 3) de vértices no sistema do mundo.

    Returns:
        np.ndarray: Array (N, 3) de vértices no sistema da câmera.
    """
    # 1. Translação: move a origem do mundo para a posição da câmera
    trans = V - C
    # 2. Rotação: alinha os eixos do mundo com os eixos da câmera
    # A rotação é feita multiplicando pela transposta da matriz R
    return (R @ trans.T).T

# Transforma cada objeto da lista 'objetos' (em coord. do mundo) para coord. da câmera
objetos_camera = []
for nome, verts, faces, cor, alpha in objetos:
    # Aplica a transformação para cada conjunto de vértices do objeto
    verts_cam = world_to_camera(verts)
    # Adiciona o objeto transformado à nova lista 'objetos_camera'
    # Mantém nome, faces, cor e alpha. O alpha é redefinido para 1.0 aqui.
    objetos_camera.append((nome, verts_cam, faces, cor, 1.0))

# --- CÁLCULO DOS LIMITES PARA O GRÁFICO NO SISTEMA DA CÂMERA ---
# Coleta todos os vértices transformados para calcular os limites dinamicamente
all_points_cam = np.vstack([obj[1] for obj in objetos_camera])

# Encontra os valores mínimo e máximo para cada eixo (X, Y, Z)
xmin_c, ymin_c, zmin_c = np.min(all_points_cam, axis=0)
xmax_c, ymax_c, zmax_c = np.max(all_points_cam, axis=0)

# Adiciona uma margem (padding) aos limites calculados
pad_c = np.max([xmax_c - xmin_c, ymax_c - ymin_c, zmax_c - zmin_c]) * 0.1

# Aplica o padding aos limites
xmin_c -= pad_c
ymin_c -= pad_c
zmin_c -= pad_c
xmax_c += pad_c
ymax_c += pad_c
zmax_c += pad_c


# --- PLOTAGEM NO SISTEMA DE COORDENADAS DA CÂMERA ---
# Cria a figura e o subplot 3D para o sistema da câmera
fig = plt.figure(figsize=(10, 10)) # Tamanho da figura ajustado para um único plot
ax2 = fig.add_subplot(111, projection='3d') # Usa 111 para um único subplot

# Itera sobre os objetos transformados para plotá-los no sistema da câmera
for nome, verts_cam, faces, cor, alpha in objetos_camera:
    # Verifica se o objeto tem faces (polígonos)
    if faces is not None and len(faces) > 0:
        # Cria uma coleção de polígonos 3D a partir dos vértices e faces
        poly3d = [[verts_cam[i] for i in face] for face in faces]
        collection = Poly3DCollection(poly3d, facecolors=cor, edgecolors="k", linewidths=0.25, alpha=1.0)
        # Adiciona a coleção ao subplot
        ax2.add_collection3d(collection)
    else:
        # Se não tem faces (é uma linha), plota a linha diretamente
        ax2.plot(verts_cam[:, 0], verts_cam[:, 1], verts_cam[:, 2], color=cor, linewidth=3)

# Plota a origem do sistema do mundo no sistema da câmera
# Transforma o ponto (0, 0, 0) do mundo para o sistema da câmera
origem_mundo_cam = world_to_camera(np.array([[0, 0, 0]]))[0]
ax2.scatter(
    origem_mundo_cam[0], origem_mundo_cam[1], origem_mundo_cam[2],
    color='black', s=180, label='Origem do Mundo', # Ponto preto para a origem do mundo
    edgecolors='white', linewidths=2, zorder=9999, depthshade=False
)

# Plota a origem do sistema da câmera (sempre em (0, 0, 0) nesse sistema)
ax2.scatter(
    0, 0, 0, color='red', s=180, label='Origem da Câmera', # Ponto vermelho para a origem da câmera
    edgecolors='white', linewidths=2, zorder=10000, depthshade=False
)

# Define os limites dos eixos do gráfico com base nos limites calculados
ax2.set_xlim(xmin_c, xmax_c)
ax2.set_ylim(ymin_c, ymax_c)
ax2.set_zlim(zmin_c, zmax_c)

# Define os rótulos dos eixos e o título do gráfico
ax2.set_xlabel('X_cam')
ax2.set_ylabel('Y_cam')
ax2.set_zlabel('Z_cam')
ax2.set_title('Sistema de Coordenadas da Câmera')

# Define a vista inicial (ângulo de elevação e azimute)
ax2.view_init(elev=28, azim=37)

# Adiciona a legenda ao gráfico
ax2.legend(loc='upper left')

# Ajusta o layout para evitar sobreposição
plt.tight_layout()

# Mostra o gráfico
plt.show()
```

Faça uma transformação de projeção em perspectiva dos sólidos contidos no volume de visão e, para isto, projete os sólidos em 2 dimensões na janela de projeção. Cada sólido deve conter arestas com mesma cor, mas sólidos diferentes devem ter cores diferentes.

a. Apresente tais objetos em 2D.

```python
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
import matplotlib.pyplot as plt
import numpy as np
import os

# -------- PROJEÇÃO EM PERSPECTIVA 2D --------
# Função para projetar vértices 3D no espaço da câmera para 2D
def project_perspective(vertices, d=1.5):
    """
    Projeta vértices 3D para 2D usando projeção em perspectiva.

    Args:
        vertices (np.ndarray): Array (N, 3) de vértices no espaço da câmera.
        d (float): Distância focal (distância do plano de projeção).

    Returns:
        np.ndarray: Array (N, 2) de pontos projetados em 2D.
    """
    X = vertices[:, 0]
    Y = vertices[:, 1]
    Z = vertices[:, 2]
    # Evita divisão por zero para vértices no plano Z=0
    Z = np.where(Z == 0, 1e-6, Z)
    # Aplica a fórmula da projeção em perspectiva
    x_proj = d * X / -Z
    y_proj = d * Y / -Z
    return np.stack([x_proj, y_proj], axis=1)

# Cria a figura e o eixo para o plot 2D
fig, ax = plt.subplots(figsize=(8, 8))
patches = [] # Lista para armazenar os polígonos (faces)
colors = []  # Lista para armazenar as cores dos polígonos

# Itera sobre cada objeto na lista de objetos_camera
for nome, verts_cam, faces, cor, alpha in objetos_camera:
    # Projeta os vértices do objeto para 2D
    verts_2d = project_perspective(verts_cam, d=1.5)

    # Verifica se o objeto tem faces definidas (não é uma linha)
    if faces is not None and len(faces) > 0:
        # --- ORDENAR FACES PELO Z MÉDIO (Painter's Algorithm - algoritmo do pintor) ---
        # Cria uma lista de tuplas (z_medio, face) para ordenar as faces
        face_z = []
        for face in faces:
            # Calcula o Z médio de cada face no espaço da câmera
            z_med = np.mean(verts_cam[np.array(face), 2])
            face_z.append((z_med, face))
        # Ordena as faces pelo Z médio (do mais distante para o mais próximo)
        face_z.sort(key=lambda x: x[0])

        # Itera sobre as faces ordenadas e cria os polígonos para plotagem
        for _, face in face_z:
            # Obtém os pontos 2D correspondentes aos vértices da face
            pts = verts_2d[np.array(face)]
            # Cria um objeto Polygon para a face
            poly = Polygon(pts, closed=True)
            # Adiciona o polígono e a cor à lista
            patches.append(poly)
            colors.append(cor)
    else:
        # Para objetos sem faces (tratados como linhas)
        # Plota a linha conectando os pontos projetados em 2D
        ax.plot(verts_2d[:, 0], verts_2d[:, 1], color=cor, linewidth=2)

# Cria uma coleção de patches para plotar todos os polígonos de forma eficiente
collection = PatchCollection(
    patches, facecolors=colors, edgecolors='k', linewidths=0.5, alpha=0.65
)
# Adiciona a coleção de polígonos ao eixo do plot
ax.add_collection(collection)

# Configura a proporção dos eixos para ser igual (para evitar distorção)
ax.set_aspect('equal')
# Define o título e as legendas dos eixos
ax.set_title("Projeção 2D dos Sólidos", fontsize=15)
ax.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=1.0)
ax.set_xlabel('x_proj', fontsize=12)
ax.set_ylabel('y_proj', fontsize=12)

# --- Ajustar limites do plot ---
# Coleta todos os pontos projetados para determinar os limites
all_proj = []
for nome, verts_cam, faces, cor, alpha in objetos_camera:
    verts_2d = project_perspective(verts_cam, d=1.5)
    all_proj.append(verts_2d)
all_proj = np.vstack(all_proj)
# Calcula um buffer para adicionar margem aos limites
buffer_x = 0.08 * (all_proj[:, 0].max() - all_proj[:, 0].min())
buffer_y = 0.08 * (all_proj[:, 1].max() - all_proj[:, 1].min())
# Define os limites do eixo X e Y com o buffer
ax.set_xlim(all_proj[:, 0].min() - buffer_x, all_proj[:, 0].max() + buffer_x)
ax.set_ylim(all_proj[:, 1].min() - buffer_y, all_proj[:, 1].max() + buffer_y)

# Ajusta o layout para evitar sobreposição de elementos
plt.tight_layout()

# ---- SALVAR NA PASTA 'imagens/' ----
# Cria a pasta 'imagens' se ela não existir
caminho_pasta = os.path.join(os.getcwd(), "imagens")
os.makedirs(caminho_pasta, exist_ok=True)
# Define o caminho completo para o arquivo de imagem
caminho_arquivo = os.path.join(caminho_pasta, "projecao_2d_solidos.png")
# Salva a figura
plt.savefig(caminho_arquivo, bbox_inches="tight")

# Mostra a figura
plt.show()
```

# Questão 05

Rasterize os objetos (poligonos) em pelo menos 3 resoluções diferentes.

```python
# Rasteirização
 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path
from matplotlib.colors import to_rgb
from skimage.draw import line as draw_line
import os

def rasterize_scene(objetos_camera, res=600, d=1.5, buffer_factor=0.08, save_path=None):
        """
        Rasteriza uma cena 3D (em coordenadas de câmera) para uma imagem 2D.

    Args:
        objetos_camera (list): Lista de objetos na cena da câmera. Cada objeto
                                é uma tupla (nome, vertices_cam, faces, cor, alpha).
                                vertices_cam: array (N, 3) de vértices no espaço da câmera.
                                faces: array (M, 3) de índices de vértices para as faces, ou None para linhas.
                                cor: string com o nome da cor.
                                alpha: transparência (não usado diretamente na rasterização atual).
        res (int): Resolução da imagem de saída (largura e altura).
        d (float): Distância focal para a projeção em perspectiva.
        buffer_factor (float): Fator para adicionar margem aos limites da imagem.
        save_path (str, optional): Caminho para salvar a imagem rasterizada.
                                   Se None, a imagem não é salva.
    """
    # 1. Projeta todos os objetos para obter limites globais na projeção
    all_proj = []
    for nome, verts_cam, faces, cor, alpha in objetos_camera:
        verts_2d = project_perspective(verts_cam, d=d)
        all_proj.append(verts_2d)
    all_proj = np.vstack(all_proj)

    # 2. Define resolução e normaliza os pontos 2D para a imagem
    min_xy = all_proj.min(axis=0)
    max_xy = all_proj.max(axis=0)
    buffer = buffer_factor * (max_xy - min_xy)
    min_xy -= buffer
    max_xy += buffer

    def normalize_to_image(xy):
        return ((xy - min_xy) / (max_xy - min_xy + 1e-8)) * (res - 1)

    # 3. Inicializa imagem branca
    image = np.ones((res, res, 3), dtype=np.float32)

    # 4. Rasterização com preenchimento e contornos
    for nome, verts_cam, faces, cor, alpha in objetos_camera:
        verts_2d = project_perspective(verts_cam, d=d)
        verts_2d_img = normalize_to_image(verts_2d)
        verts_2d_img[:, 1] = res - 1 - verts_2d_img[:, 1]  # Inverte Y

        if faces is not None and len(faces) > 0:
            # --- Ordenar as faces pelo Z médio no espaço da câmera ---
            face_z = []
            for face in faces:
                z_med = np.mean(verts_cam[np.array(face), 2])
                face_z.append((z_med, face))
            face_z.sort(key=lambda x: x[0])  # Ordenando pelo z_med (1° elemento)

            for _, face in face_z:
                pts = verts_2d_img[np.array(face)]

                path = Path(pts)
                x_min, y_min = np.floor(pts.min(axis=0)).astype(int)
                x_max, y_max = np.ceil(pts.max(axis=0)).astype(int)

                x_min = max(x_min, 0)
                y_min = max(y_min, 0)
                x_max = min(x_max, res - 1)
                y_max = min(y_max, res - 1)

                grid_x, grid_y = np.meshgrid(
                    np.arange(x_min, x_max + 1),
                    np.arange(y_min, y_max + 1)
                )
                grid_points = np.stack([grid_x.ravel(), grid_y.ravel()], axis=1)

                mask = path.contains_points(grid_points)
                mask = mask.reshape((y_max - y_min + 1, x_max - x_min + 1))

                color = np.array(to_rgb(cor)).reshape(1, 1, 3)
                image[y_min:y_max+1, x_min:x_max+1][mask] = color

                # --- Desenhar arestas (bordas pretas) ---
                pts_int = pts.astype(int)
                n = len(pts_int)
                for i in range(n):
                    x0, y0 = pts_int[i]
                    x1, y1 = pts_int[(i + 1) % n]
                    rr, cc = draw_line(y0, x0, y1, x1)  # (linha, coluna)
                    valid = (rr >= 0) & (rr < res) & (cc >= 0) & (cc < res)
                    image[rr[valid], cc[valid]] = [0, 0, 0]  # preto

        else:
            # Rasteriza linha reta (em pontos discretos)
            for i in range(len(verts_2d_img) - 1):
                p1 = verts_2d_img[i]
                p2 = verts_2d_img[i + 1]
                n_points = int(np.linalg.norm(p2 - p1) * 2)
                xs = np.linspace(p1[0], p2[0], n_points).astype(int)
                ys = np.linspace(p1[1], p2[1], n_points).astype(int)
                rgb = np.array(to_rgb(cor), dtype=np.float32)
                for x, y in zip(xs, ys):
                    if 0 <= x < res and 0 <= y < res:
                        image[y, x] = rgb

    # --- Plot e salvamento ---
    plt.figure(figsize=(8, 8))
    plt.imshow(image)
    plt.axis("off")
    plt.title(f"Rasterização {res}x{res}", fontsize=16)

    if save_path:
        caminho_pasta = os.path.dirname(save_path)
        if caminho_pasta and not os.path.exists(caminho_pasta):
            os.makedirs(caminho_pasta)
        plt.savefig(save_path, bbox_inches="tight", pad_inches=0)

    plt.show()
```
